Matploblib:


*   Useful for making plots





```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100) #equally spaced values between 1 -10 i.e 100 values
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sin wave
plt.plot(x, y)
plt.show()
```

```python
#cosine wave
plt.plot(x, z)
plt.show()
```

```python
# Add title x-axis and y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sine wave')
plt.show()
```

```python
# plot a parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+') 
# r - red
# + - symbol
plt.show()
```

```python
plt.plot(x,y, 'g.') 
plt.show()
```

```python
plt.plot(x,y, 'rx') 
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r.')
plt.show()
```

Bar plot  - Data visualization

```python
fig = plt.figure() #creates empty plot
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'spanish', 'latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

Pie chart

```python
fig1 = plt.figure() #creates empty plot
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'spanish', 'latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 *np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, c =z,cmap='Blues')
plt.show()
```
